# Pivot Tables

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Salaries.csv')
```

```python
df.head(5)
```

```python
pd.pivot_table(data=df,values='BasePay',index='EmployeeName',aggfunc='mean')
```

```python
pd.pivot_table(data=df,values='JobTitle',index='TotalPay',aggfunc='sum')
```

```python
pd.pivot_table(data=df,values='JobTitle',index='OvertimePay',aggfunc='max')
```

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])
```

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K5'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})    
```

```python
left
```

```python
right
```

```python
pd.merge(left=left,right=right,how='inner',on='key')
```

```python
pd.merge(left=left,right=right,how='outer',on='key')
```

```python
pd.merge(left=left,right=right,how='left',on='key')
```

```python
pd.merge(left=left,right=right,how='right',on='key')
```

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
pd.merge(left=left,right=right,how='inner',on=['key1','key2'])
```

```python
pd.merge(left=left,right=right,how='outer',on=['key1','key2'])
```

```python
pd.merge(left=left,right=right,how='left',on=['key1','key2'])
```

```python
pd.merge(left=left,right=right,how='right',on=['key1','key2'])
```
