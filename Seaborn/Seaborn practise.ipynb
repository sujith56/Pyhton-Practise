```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
dataset = np.random.randint(-25,25,25)
```

```python
dataset
```

```python
sns.rugplot(dataset)
```

```python
x_min = dataset.min()
x_max = dataset.max()
```

```python
x_axis = np.linspace(x_min,x_max,100)
```

```python
x_axis
```

```python
bandwidth = ((4*dataset.std()**4)/(5*len(dataset)))**2
```

```python
bandwidth
```

```python
kernel_list=[]
for data_point in dataset:

kernel = stats.norm(dataset,bandwidth).pdf(x_axis)
kernel_list.append(kernel)
kernel = kernel/kernel.max()
kernel = kernel* .4
plt.plot(x_axis,kernel,color='grey',alpha=0.8)
plt.ylim(0,1)
```

```python
sum_of_kde = np.sum(kernel_list,axis=0)
```

```python
fig = plt.plot(x_axis,sum_of_kde,color='indianred')
sns.rugplot(dataset,c = 'indianred')
plt.yticks([])
plt.suptitle("Sum of the Basis Functions")
```
