# Adding Legends

```python
import numpy as np 
from matplotlib import pyplot as plt 

x = np.arange(0,10) 
y = x ^ 2 
z = x ^ 3

# Labeling the Axes and Title
plt.title("Graph Drawing") 
plt.xlabel("Time") 
plt.ylabel("Distance") 

# Adding Legends
plt.plot(x,y,label='Race1')
plt.plot(x,z,label='Race2')
plt.legend()
```


# Legend at different location with shadow enabled and fontsize set to large

```python
x = np.arange(0,10) 
y = x **2 
z = x ** 3
t = x ** 4 

# Labeling the Axes and Title
plt.title("Graph Drawing") 
plt.xlabel("Time") 
plt.ylabel("Distance") 

# Adding Legends
plt.plot(x,y,label='Race1')
plt.plot(x,z,label='Race2')
plt.plot(x,t,label='Race3')


plt.legend(loc='best',shadow=True,fontsize='larger')
```

# Bar Chart

Syntax:

    matplotlib.pyplot.bar(x, height, width=0.8, bottom=None, *, align='center', data=None, **kwargs)
  
  Use the **`bar()`** function to make bar charts, which includes customizations such as error bars:
  
  

# Vertical Bar Chart:

```python
objects = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
performance = [10,8,6,4,2,1]
 
plt.bar(objects, performance, width = 0.2, alpha=0.9)
plt.xlabel('Company')
plt.ylabel('Usage')
plt.title('Programming language usage')
```

# Horizontal bar chart:

```python
objects = ('welcome to Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
performance = [10,8,6,4,2,1]
 
plt.barh(objects, performance, align='center', alpha=0.5)
plt.xlabel('Usage')
plt.title('Programming language usage - pythonspot.com')
```

# Scatter Plot

Syntax:
    
    matplotlib.pyplot.scatter(x, y, s=None, c=None, marker=None, cmap=None, norm=None, vmin=None, vmax=None, alpha=None, linewidths=None, verts=None, edgecolors=None, *, data=None, **kwargs)[source]
    
    Reference: https://matplotlib.org/api/_as_gen/matplotlib.pyplot.scatter.html#matplotlib.pyplot.scatter

```python
import numpy as np
import matplotlib.pyplot as plt

N = 20
x = np.random.rand(N)
y = np.random.rand(N)
plt.scatter(x, y, c='r', alpha=0.8)
plt.title('Scatter plot')
plt.xlabel('x')
plt.ylabel('y')
```

# Pie Chart


    matplotlib.axes.Axes.pie
    Axes.pie(x, explode=None, labels=None, colors=None, autopct=None, pctdistance=0.6, shadow=False, labeldistance=1.1, startangle=None, radius=None, counterclock=True, wedgeprops=None, textprops=None, center=(0, 0), frame=False, rotatelabels=False, *, data=None)

# Creating multi plots on same canvas

```python
x = np.arange(0,10)
y = x**2


# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,2,1)
plt.plot(x, y, 'r--')# More on color options later
plt.title('Plot1')
plt.xlabel('x_1')
plt.ylabel('y_1')

plt.subplot(1,2,2)
plt.plot(y, x, 'g*-')
plt.title('Plot2')
plt.xlabel('x_2')
plt.ylabel('y_2')
```
